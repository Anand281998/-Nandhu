```python
#importing the libraries
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
#importing the data set
z = pd.read_csv('http://ce.sharif.edu/courses/85-86/1/ce925/assignments/files/assignDir2/churn.txt')
z
```

```python
# correlation between the attributes
z.corr(method = 'pearson')
```

```python
#importing data into feature and label variables
x = z.iloc[:,1:20].values
y = z.iloc[:,20].values
```

```python
#converting the string data into binary form
le = LabelEncoder()
x[:,2] = le.fit_transform(x[:,2])
x[:,3] = le.fit_transform(x[:,3])
x[:,4] = le.fit_transform(x[:,4])
y = le.fit_transform(y)
```

```python
X = x
Y = y
```

```python
# splitting the data into train and test
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.3)
```

```python
# scaling all attributes
sc = StandardScaler()
x_train =sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
# creating the function
classifier = Sequential()

# creating the layers
layer1 = classifier.add(Dense(1000, kernel_initializer = 'uniform' ,input_dim = 19 ,activation = 'tanh'))
layer2 = classifier.add(Dense(500, kernel_initializer = 'uniform' , activation = 'tanh'))
layer_3 = classifier.add(Dense(250 , kernel_initializer = 'uniform' , activation = 'tanh'))
layer_4 = classifier.add(Dense(100,kernel_initializer = 'uniform' , activation = 'tanh'))
layer_5 = classifier.add(Dense(25, kernel_initializer = 'uniform' ,activation = 'tanh'))
out_layer = classifier.add(Dense(1 ,kernel_initializer = 'uniform' ,activation = 'sigmoid'))
```

```python
# optimizing the loss function using optimizer
classifier.compile(optimizer = 'Adam' , loss = 'binary_crossentropy' , metrics = ['accuracy'])

#fitting the data
fit = classifier.fit(x_train , y_train , batch_size = 100 ,epochs= 150)
```

```python
# Predicting the test data
y_pred = classifier.predict(x_test)

# Accuracy of the test prediction
print(accuracy_score(y_test , y_pred.round(),normalize = True)*100)
```
